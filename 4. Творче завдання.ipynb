```python
import pandas as pd
import matplotlib as plt
df = pd.read_csv('../../data/yulu_rental.csv')
```

Питання 1: чи є різниця між звичайними і зареєстрованими користувачами протягом робачого дня? І коли пікові показники значень?

```python
df['datetime'] = pd.to_datetime(df.datetime)
df['hour'] = df.datetime.dt.hour
```

```python
clients_by_hour_workday = df[df.workingday==1][['casual', 'registered', 'hour']].groupby('hour').sum()
clients_by_hour_workday.plot(title='Користувачі протягом робочого дня');
```

Питання 2: чи є різниця між звичайними і зареєстрованими користувачами протягом вихідного дня? І коли пікові показники значень?

```python
clients_by_hour_holiday = df[df.workingday==0][['casual', 'registered', 'hour']].groupby('hour').sum()
clients_by_hour_holiday.plot(title='Користувачі протягом вихідного дня');
```
